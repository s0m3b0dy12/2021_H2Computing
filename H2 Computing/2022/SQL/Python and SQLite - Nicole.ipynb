**Exercise 2**

Write a program that read the file books.txt and store the name of book in the BOOK table and name of the author in the AUTHOR table.

If there are no author information, replace the author as "unknown".

````python
import sqlite3 
connection = sqlite3.connect("library.db")

f = open("books.txt",'r')
for line in f:
    line = line.strip('\n').strip('"').split('" by ')
    name = line[0]
    author = line[1]
    
    author_table = connection.execute(f"SELECT NAME FROM AUTHOR WHERE NAME=\"{author}\"").fetchall()
    if not author_table:
        connection.execute(f"INSERT INTO AUTHOR(NAME) VALUES(\"{author}\")")
        
    elif author == '':
        connection.execute("INSERT INTO AUTHOR(NAME) VALUES(\"Unknown\")")
    
    author_id=connection.execute(f"SELECT ID FROM AUTHOR WHERE NAME=\"{author}\"").fetchall()
    for row in author_id:
        k = row[0]
        connection.execute(f"INSERT INTO BOOK(NAME, AUTHORID) VALUES(\"{name}\", \"{k}\")")
        
        
connection.commit()
connection.close()
````
